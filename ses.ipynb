## py3Dmol rendering

```python
SELECTED_MOLECULE = MOLECULE_NAMES[0]
SELECTED_MOLECULE
```

```python
analytic_py3dmol_view = render_py3dmol_analytic_ses_points(SELECTED_MOLECULE, point_area=0.05)
```
