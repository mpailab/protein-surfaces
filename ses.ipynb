# SES points with py3Dmol

Минимальный ноутбук: определения молекул и py3Dmol-viewer для SES-точек, ребер локального surface-графа и нормалей к поверхности.


```python
from pathlib import Path
import sys

import torch

ROOT = Path.cwd()
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from molecule.examples import MOLECULES, MOLECULE_NAMES, get_molecule
from molecule.visualize import (
    render_py3dmol_analytic_ses_points,
    render_py3dmol_points,
    render_py3dmol_sdf_ses_points,
    render_py3dmol_tiled_analytic_ses_points,
    sample_display_points,
)

torch.set_default_dtype(torch.float64)
```

## Molecules

```python
# Molecule definitions live in src/molecule/examples.py
MOLECULE_NAMES
```

## py3Dmol rendering

```python
SELECTED_MOLECULE = MOLECULE_NAMES[0]
SHOW_SES_EDGES = True
SHOW_SES_NORMALS = True
SELECTED_MOLECULE
```

```python
analytic_py3dmol_view = render_py3dmol_analytic_ses_points(
    SELECTED_MOLECULE,
    point_area=0.05,
    show_edges=SHOW_SES_EDGES,
    show_normals=SHOW_SES_NORMALS,
)
```

```python
sdf_py3dmol_view = render_py3dmol_sdf_ses_points(
    SELECTED_MOLECULE,
    m=128,
    smoothness=0.3,
    max_grid_points=200_000,
    show_edges=SHOW_SES_EDGES,
    show_normals=SHOW_SES_NORMALS,
)
```

```python
tiled_analytic_py3dmol_view = render_py3dmol_tiled_analytic_ses_points(
    SELECTED_MOLECULE,
    point_area=0.05,
    tile_size="auto",
    tile_overlap="auto",
    atom_density_scale=1.55,
    pair_density_scale=1.55,
    probe_density_scale=1.55,
    dedup_tolerance=0.05,
    exact_accessibility=False,
    max_grid_points=500_000,
    max_probe_triples=5_000_000,
    color_by="block_type",
    show_edges=SHOW_SES_EDGES,
    show_normals=SHOW_SES_NORMALS,
)
```
